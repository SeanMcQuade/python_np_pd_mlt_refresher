```python
## First load all possible letters into vectors, a word is one selection from each of 6--8 vectors each vector has 8 elements.
```

```python
RedP_words = ([["l","s","m","w","b","c","p","t"],["e","i","o","u","h","r","t","a"],
               ["n","t","s","o","i","v","l","r"],["t","h","s","i","k","v","a","d"],
               ["n","l","r","c","s","a","e","o"],["r","w","t","e","y","n","d","g"]])

Lost_eyes = ([["f","r","l","h","c","s","v","p"],["a","e","i","o","u","h","l","r"],
               ["r","p","g","a","o","c","m","t"],["s","t","r","d","l","e","i","u"],
               ["e","i","u","t","r","l","h","a"],["e","z","h","l","t","n","r","s"],["r","n","y","s","g","h","d","e"]])

Sin_infinity = ([["h","l","m","n","r","t","d","g"],["a","e","i","o","u","h","r","y"],
               ["f","n","s","o","m","r","a","u"],["i","o","u","p","g","t","a","e"],
               ["n","o","r","s","t","c","g","l"],["i","o","h","n","l","t","a","e"],
               ["t","o","l","u","c","e","d","s"],["y","h","e","s","n","r","k","t"]])
```

The following cell builds all possible strings from Red_P and saves them into the Red_P_word_list list.

```python
import numpy as np
Red_P_word_list = []
for i1 in np.arange(8): 
    for i2 in np.arange(8):
        for i3 in np.arange(8):
            for i4 in np.arange(8):
                for i5 in np.arange(8):
                    for i6 in np.arange(8):
                        candidateWord = []
                        letter_indices = [i1,i2,i3,i4,i5,i6]
                        for i in np.arange(6):                            
                            candidateWord.append(RedP_words[i][letter_indices[i]])
                        #print(candidateWord[0]+candidateWord[1]+candidateWord[2]+candidateWord[3]+candidateWord[4]+candidateWord[5])
                        Red_P_word_list.append(''.join(candidateWord))  #''.join turns the 'l', 'e', ... into 'le...'
```

Red_P_word_list is checked against an English (US) dictionary. Only the strings in the intersection of list and dictionary are printed.

```python
import enchant
Red_P_Eng_word = []
d = enchant.Dict("en_US")
#print("The number of words is", len(Red_P_word_list)) #print the number of words
for i in np.arange(len(Red_P_word_list)):
    if d.check(Red_P_word_list[i]): # using the not operator to print the false
        Red_P_Eng_word.append(Red_P_word_list[i])
        #print(Red_P_word_list[i])
print("The number of words is", len(Red_P_Eng_word)) #print the number of words
```

```python
import numpy as np
Lost_eyes_word_list = []
for i1 in np.arange(8):
    for i2 in np.arange(8):
        for i3 in np.arange(8):
            for i4 in np.arange(8):
                for i5 in np.arange(8):
                    for i6 in np.arange(8):
                        for i7 in np.arange(8):
                            candidateWord = []
                            for i in np.arange(7):
                                letter_indices = [i1,i2,i3,i4,i5,i6,i7]
                                candidateWord.append(Lost_eyes[i][letter_indices[i]])
                            Lost_eyes_word_list.append(''.join(candidateWord))  #''.join turns the 'l', 'e', ... into 'le...'
import enchant
Lost_eyes_Eng_word = []
d = enchant.Dict("en_US")
#print("The number of words is", len(Lost_eyes_word_list))
for i in np.arange(len(Lost_eyes_word_list)):
    if d.check(Lost_eyes_word_list[i]): ## the d.check() is boolean, and can be used in the if statement directly
        #print(Lost_eyes_word_list[i])
        Lost_eyes_Eng_word.append(Lost_eyes_word_list[i])
print("The number of words is", len(Lost_eyes_Eng_word))
```
